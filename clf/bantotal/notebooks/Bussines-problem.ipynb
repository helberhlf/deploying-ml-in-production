 ### <center>Desafio de negócio</center>


A plataforma bancária Bantotal deseja automatizar o processo de elegibilidade de empréstimos com base nos detalhes do cliente fornecidos,cada vez que um cliente solicita um empréstimo a instituição financeira, diversos processos e controles internos são acionados, necessários para a avaliação da solicitação recebida. Desta forma, são analisadas manualmente muitas informações relacionadas com o perfil do cliente, destinos de crédito, atividade de trabalho, rendimentos, condições de habitação, entre outros dados demográficos. Além disso, a instituição utiliza os chamados bureaus de crédito para conhecer o histórico de crédito do cliente, a fim de definir o perfil de crédito do cliente. Em conjunto com outros registros próprios, informações de outros créditos, grau de conformidade e comportamento dos produtos contratados, a instituição aprova um valor a ser emprestado e um prazo para sua devolução ou rejeita a solicitação.

### <center> Objetivo</center>


**O desafio consiste em criar um modelo de inteligência artificial capaz de realizar uma análise de risco para predizer se um empréstimo a um cliente deve ser (autorizado ou negado)**. Deste modo,
espera-se a utilização de um modelo de Machine Learning capaz de realizar uma classificação. O modelo deverá ser treinado com os conjuntos de dados disponibilizado pelo organização, contendo dados de clientes de bancos, como dados demográficos, sobre suas contas e sobre o empréstimo que querem realizar.

### <center> Desenvolvimento </center>
O desafio consiste na elaboração de um modelo de Machine Learning para predição de risco de empréstimo, baseado em informações bancárias. 

    O modelo deve receber como entrada os seguintes dados:

###  Aprendizagem Supervisionada e Não Supervisionada


Os termos **supervisionado e não supervisionado foram herdados do campo de Aprendizado de Máquina, em inglês Machine Learning - ML ou AM.** Metaforicamente, um professor "supervisiona “o estudante fornecendo, cuidadosamente, informações alvo, junto com um conjunto de exemplos. Uma tarefa de aprendizagem não supervisionada pode implicar no mesmo conjunto de exemplos, mas não incluiria as informações alvo (destino). O estudante não receberia informação sobre o objetivo do aprendizado, mas ficaria livre para tirar as próprias conclusões sobre o que os exemplos têm em comum **(FAWCETT; PROVOST, 2016)**.

Os sistemas de Aprendizado de Máquina podem ser classificados de acordo com a quantidade e o tipo de supervisão que recebem durante o treinamento. Existem quatro categorias principais de aprendizagem: supervisionado, não supervisionado, semi-supervisionado e por reforço **(GÉRON, 2021)**
    
    
 ###  Modelos, Indução e Previsão

Em geral, **um modelo é uma representação simplificada da realidade criada para servir a um propósito.** Assim, ele se baseia-se em algumas simplificações de suposições sobre o que é importante e o que não é específico ou às vezes com base em restrições de informações ou tratabilidade. Por exemplo, um mapa é um modelo do mundo físico. Ele abstrai uma enorme quantidade de informações que o cartógrafo considera irrelevantes para sua finalidade. Ele preserva e, às vezes, simplifica ainda mais, as informações relevantes.**(FAWCETT ; PROVOST, 2016)**.


Em **Data Science**, um modelo preditivo é uma fórmula para estimar o valor **desconhecido de interesse**: **o alvo**. A fórmula pode ser matemática ou pode ser uma declaração lógica, como uma regra. Muitas vezes, é um híbrido ambos os dois. Assim sendo, dada nossa divisão de mineração supervisionada de dados em **classificação e regressão**, vamos considerar modelos de classificação (modelos que prevê a estimativa de probabilidade de classe) (FAWCETT ; PROVOST, 2016).

<font color='red'>Lembre-se de que os modelos preditivos se concentram em estimar o valor da variável alvo de interesse.</font>

###  Termo-chave: previsão

No uso comum, **previsão significa antecipar um evento futuro**. Em **Data Science**, de forma mais geral, previsão significa **estimar um valor desconhecido**. Esse valor poderia ser algo no futuro (no uso comum, previsão verdadeira), mas também pode ser algo no presente ou no passado. De fato, como a mineração de dados geralmente lida com dados históricos, os modelos costumam ser construídos e testados por meio do **uso de eventos do passado**. Modelos preditivos para pontuação de crédito estimam a probabilidade de um potencial cliente ser inadimplente. Modelos preditivos para filtragem de spam estimam se determinado e-mail é spam. Modelos preditivos para a detecção de fraude julgam se uma conta foi defraudada. Em poucas palavras, o segredo é que o modelo seja destinado para **estimar um valor desconhecido** (FAWCETT ; PROVOST, 2016).


###  Termo-chave: Indução e Dedução

A indução pode ser comparada com a **dedução**. Porém, começa com regras gerais e fatos específicos, e cria outros fatos específicos a partir deles. A utilização dos nossos modelos pode ser considerada um procedimento de dedução **(probabilística)** (FAWCETT ; PROVOST, 2016).

Além disso, a criação de modelos a partir de dados é conhecida como **modelo de indução.** A indução é um termo da filosofia que se refere a generalização a partir de casos específicos até regras gerais (ou leis ou verdades). Nossos modelos são regras gerais no sentido estatístico (eles normalmente não se mantêm 100% das vezes; com frequência, nem chegam perto), e o procedimento que cria o modelo a partir de dados é chamado **algoritmo de indução ou aprendiz** (FAWCETT and PROVOST, 2016).
