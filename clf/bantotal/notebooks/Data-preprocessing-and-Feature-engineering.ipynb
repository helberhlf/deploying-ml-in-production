### 3.1. Pipelines & Combination of attributes

```python
"""
Creating a pipeline to preprocess the numeric attributes
                        and 
Creating a pipeline to preprocess the categorical attributes:
"""
num_pipe = Pipeline([
        ("imputer", SimpleImputer(strategy="median")),
        #("norm", Normalizer()),
        #("min_max", MinMaxScaler()),
        
])
'''
lb_pipe = Pipeline([
    ('lb', transformer.CustomLabelEncoder()), 
])
'''
cat_pipe = Pipeline([  
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ('ohe', OneHotEncoder(handle_unknown='ignore',sparse=False)),
])

# Select numeric attributes and  Select categorical attributes
num_attribs = ["installment_percent","loan_amount","payment_term"]
#lb_attribs  = ["installment_plans","others_on_loan"]
cat_attribs = ["loan_purpose","installment_plans","others_on_loan"]

# Create a complete data preprocessing pipeline using ColumnTransformer and transformer pipelines. 
#So, let's bring the numeric and categorical pipelines together:
preprocess_pipeline = ColumnTransformer([
        ("num_pipe",num_pipe,num_attribs),
        #("lb_pipe",lb_pipe,lb_attribs),
        ("cat_pipe",cat_pipe,cat_attribs),
],
        remainder='passthrough'
)
# View
preprocess_pipeline

# Fit the data, that is, transform
df_clean = preprocess_pipeline.fit_transform(df_drop_dup)

# Extracting column names
cols=transformer.get_column_names_from_ColumnTransformer(preprocess_pipeline)

# Creating a new dataframe to check if the necessary transformations were transformed successfully
df_new = pd.DataFrame(df_clean, columns=cols)
df_new.rename(columns = {'6':'allow'}, inplace = True)

# Calculanting value per installment
df_new['value_per_installment'] = df_new.loan_amount / df_new.installment_percent

# Calculating year
# years =  number of days // 365
df_new['years'] = df_new.payment_term.map(lambda x : x // 365 )

# Calculating months
# months = (number of days - years * 365) // 30
df_new['months'] = (df_new.payment_term - df_new.years * 365) // 30 

# Drop_duplicates se existir
df_new=df_new.drop_duplicates(keep='last')

# View 
df_new.head()
```

### 3.3. What is interest rate

Segundo (JEAN CAVALEIRO, 2020) defende que  taxa de juros é o coeficiente que determina o valor do juro, a razão entre os juros recebidos (ou pagos) e o capital incicial aplicado (ou emprestado). Elas se referem sempre a uma undiade de tempo (dia, mês, ano e etc) e podem ser representado  equivalemente de duas maneiras: **taxa unitária (fração decimal) e taxa percentual; uma é fruto da outra, não são duas coisas diferentes, e sim uma extensão da aplicação.**

Então, partir dos conceitos gerais de **Matemática Financeira**, vamos conceituar os rercusos necessários para entender um pouco mais sobre este assunto, conforme (JEAN CAVALEIRO, 2020) descreve nas seções subsequentes:
 
#### Capital

> É o valor principal de uma operação, ou seja, do dinheiro em um momento inicial.

Por exemplo: fez-se uma aplicação no valor de Rs 10.000,00 por 5 meses, a uma taxa de juros de Rs de 1% ao mês. **O capital é de R$ 10.000,00**


#### Montante

> Em resumo, podemos entendê-lo como o valor do dinheiro do futuro.

Por exemplo: apliquei R$ 10.000,00 na caderneta de poupança e fui viajar para Europa, passei 2 anos fora. Qual é o montante da minha aplicação quando voltei ao Brasil?

#### Juros

> É a correção  monetária em espécie ou valor acrescido pela taxa de juros. Assim como, se tenho R$ 1.000,00 e pela aplicação feita por ano saquei Rs 1.500,00, posso dizer que recebi um juros de Rs 500,00. Pode ser **simples ou composto**

#### Período 

São os prazos envolvidos na operação financeira. Podem ser expressos em dia, semana, mês, semestre, ano; mas o que importa é que temos de considerar um regra: devem constar, de um problema de Matemática Financeir, todasm as informações de taxa e período na mesma menção de tempo. **Vejamos um exemplo. Fez-se uma aplicação de R$ 1.000,00 por 6 meses, a uma taxa de 2% ao bimestre. Devo  transformar a taxa em mês ou período em bimestre, ficando o valor aplicado por 3 bimestres, que é o mesmo que 6 meses**


#### Investimento 

> Operação financeira em que se faz aplicação de um valor e espera recebê-lo acrescido dos juros incorridos no período.


#### Empréstimo

> Operação financeira na qual se buscam recurso no mercado para fazer frente às necessidades das mais variadas espécies.


#### Amortização

> Antecipação de pagamentos de operação de financiamentos, na qual se fazem  necessários os conceitos de valor atual e futuro

### Juros simples

> A capitalização ocorre somente sobre o valor principal, ou capital inicial

### Juros compostos

> A capitalização ocorre sobre o valor principal e sobre  os juros incorridos; é chamando **juros sobre juros.**

### 3.3.1. Interest capitalization regime

#### Regime de capitalização  simples

> Observase-que (JEAN CAVALEIRO, 2020) compara-se a uma progressão aritmética, isto é, o juro **cresce de forma linear** ao longo do tempo. Os juros incidem somente sobre o capital inicial da operação, e não sobre o acumulativo.


Lembrete do autor!

> O regime de juros simples tem como particularidade a incidência  dos juros sobre o valor principal do empréstimo, isto é, os cálculos dos montantes (capital + juros) serão realizados com referência ao valor principal, indepedente do período. Sobre os juros gerados em cada período, não incidirão novos juros

É importante ressaltar que valor principal ou simplesmente principal ou capital é o valor inicial emprestado ou aplicado antes de somarmos os juros.


Vejamos algumas opções:

Em juros simples podemos utilizar as seguintes fórmulas:

- Para cálculo dos juros:


$$J = C * i * n$$

-  Para cálculo do capital:

$$ C = J / i * n $$


-  Para cálculo da taxa de juros:

$$ i = J / c * n $$


- Para cálculo do período:

$$n = J / c * i$$

Onde que:

-  J = juros
-  i = taxa de juros
-  n = números de períodos



### 3.3.2. Montante

Além do mais, (JEAN CAVALEIRO, 2020) destaca-se um pouco mais termo-chave  **montante**, é o valor obtido através da soma do capital e dos juros produzidos em um determinado período; alguns autores, nas resoluções de exercícios, representam o montante pela letra M, e outras utilizam a abreviação FV

> Montante é o valor do dinheiro em tempo futuro.


Como já foi mencionado anteriormente, o montante pode ser obtido por meio da soma do capital com os juros

$$Montante  = capital + juros$$

Capital é um valor monetário disponível em determinado data. Sempre é considerado a data zero, **o momento em que é feita uma aplicação ou empréstimo.** Além disso, sua representação ocorre pela letra C ou pela abreviatura PV, que, do inglês, **significa valor presente.** Não importando por qual dessas representações optar, o significado é o mesmo

> Capital é o valor presente do dinheiro, o valor na data zero da operação

Mediante o exposto, os conceitos de capital e montante, é possível obtermos as seguintes fórmulas:

- Juros
  
$$J = C * i * n$$

-  Montante
  
$$M = C + j$$

Além do mais, os juros não forem expressos, a fórmula do montante será:

$$M = C(1 + i * n)$$

Capital

$$C = M/(1 + i * n)$$

```python
# Interest rate
I = 0.012

def simple_cap_regime(c,n):
    # For interest calculation
    j =  c * I * n

    """
    For capital calculation
    
    c = j /i+n
    For calculation of the interest rate
    i  = j/c*n
    
    For calculation of the period
    n = j/c*i
    
    return j, c, i , p
    """
    return j
# def acc_amount(j,c):return j+c
def value_future(c,n): return c * (1 + I*n) # O valor futuro é o valor do dinheiro após um determinado período, é o que chamamos de montante
```

```python
df_new["capitalized_interest"] = simple_cap_regime(df_new.loan_amount,df_new.months)
df_new["value_future"]         = value_future(df_new.loan_amount,df_new.months)
```

```python
# Outlier treatment use using
def outliers(x):
    """
    A função retorna um vetor booleano: True se o elemento for um outlier. Falso, caso contrário.
    Agora, para substituir os outliers superiores e inferiores, vamos escrever outra pequena função e aplicá-la em todo o dataframe:
    """
    return np.abs(x-x.median()) > 1.5 * (x.quantile(0.95) - x.quantile(0.05))

def replace(x):
    """
    Substitua os valores outliers mais altos pelo percentil 95 e os valores outliers mais baixos pelo percentil 5
    """
    out = x[outliers(x)] 
    return x.replace(to_replace=[out.min(), out.max()],
                     value=(np.percentile(x,0.05), np.percentile(x,0.95)))
```

```python
# Aplique replace() em cada coluna do dataframe
df_before= df_log1p.apply(replace,axis=1)

# Por fim, remova as linhas que contêm qualquer valores outliers pelo método mais simples IQR
df_after = df_before[~df_before.apply(outliers).any(axis=1)]

functions_aux.summary_quick_in_between_datasets(df_before,df_after)
#print('\nDifference:',abs(df_before.shape[0] - df_after.shape[0]),'drop outliers')
```
